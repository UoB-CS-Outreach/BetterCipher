1. **Substition ciphers**, where characters are transformed into other characters without following a fixed rule;

## Substitution ciphers

A general substitution cipher allows any character to be substituted for any other, rather than imposing a fixed rule such as with Ceaser ciphers. The main advantage of this is that there are hugely more possible substitutions, making a bute force attack far less effective. Specifically, if the alphabet has $n$ characters then there are $n!$ different keys, which is very large even for a relatively small alphabet. For example, there are $26$ lower case letters in the English language, giving $26! \approx 4 \times 10^{26}$ keys. If an attacker could test one billion keys a second, a brute force attack would take over six billion years to succeed!

Substitution ciphers are very straightforward:

1. choose a re-arrangement (**permutation**) of the alphabet as the key;
2. to encrypt a character, substitute it for the character at the same position in the key.

Substitution ciphers are easy to understand, have a simple implementation, and are resistant to brute force attacks. However, they are not secure because statistical properties of languages are well known, making them trivial to attack using [letter frequency analysis](https://en.wikipedia.org/wiki/Letter_frequency): find the frequence of each character in the ciphertext, and match these up against the known frequencies of the language used. The most frequent character in the ciphertext is most likely the most frequent letter in that language. In English, the most frequently used character is `e`, so the most frequent character in the ciphertext is probably `e`.

Using these statistics, you can construct partial keys which cause high-frequency clusters to appear, and then brute force the remainder of the key.

## Product ciphers

Product ciphers combine substitutions and permutations to create a resulting cipher that is significantly stronger than either a substitution or permutation alone. The simplest product cipher can be easily described: apply a subsitution cipher to the plaintext then apply a permuation cipher to the substitution cipher's output. Product ciphers are strong because they obscure the statistical properties of the plaintext since characters appear to be substituted to different characters in the ciphertext, depending on where they appear in the original text. The immediate consequence of this is that statistical attacks based on letter frequences are far less effective and other types of attacks are required. You can read more about this in the Wikipedia article on [confusion and diffusion](https://en.wikipedia.org/wiki/Confusion_and_diffusion).

Product ciphers form the basis for modern ciphers such as [the Advanced Encryption Standard (AES)](https://en.wikipedia.org/wiki/Advanced_Encryption_Standard), though they include many combinations of substitutions and permutations, as well as other operations. Understanding substitutions, permutations, and product ciphers gives you a strong foundation in the state of the art of modern encryption - for example, AES is the only publicly accessible cipher approved by the US National Security Agency (NSA) for securing top secret information.
